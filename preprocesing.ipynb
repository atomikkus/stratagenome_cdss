```python
import pandas as pd
```

```python
cnv= pd.read_csv(r"cnv_ap_matrix.csv")
cnv
```

```python
cnv_stats = cnv.sum(axis=0).reset_index().iloc[1:]
cnv_stats[['CNV', 'freq']] = cnv_stats 
cnv_stats.drop(['index', 0], axis=1, inplace=True)
```

```python
cnv_stats.sort_values('freq', ascending=False)
```

```python
combi_df = combi_df.fillna(0).drop('filename', axis =1)
```
